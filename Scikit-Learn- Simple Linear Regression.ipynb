```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng=np.random.RandomState(42)
```

```python
print(rng)
```

```python
x=10*rng.rand(50)
```

```python
x
```

```python
y=2*x-1+rng.randn(50)
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression(fit_intercept=True)
```

```python
model
```

```python
x
```

```python
x
```

```python
X=x[:,np.newaxis]
```

```python
X
```

```python
model.fit(X,y)
```

```python
model.coef_ #slope
```

```python
model.intercept_ #intercept
```

```python
xfit=np.linspace(-1,11)
```

```python
xfit
```

```python
Xfit=xfit[:,np.newaxis]
```

```python
Xfit
```

```python
yfit=model.predict(Xfit)
```

```python
plt.scatter(x,y)
plt.plot(xfit,yfit)
```
